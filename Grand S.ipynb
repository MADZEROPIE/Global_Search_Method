```python
import subprocess
str_of_r=["2.0","3.0","4.0"]

s="res"

for i in str_of_r:
    subprocess.run([r".\build\Release\builder_sample.exe",s+i,i])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

names=["Hansen","Hill","Shekel"]

print(names[0]);

for i in str_of_r:
    #print("r =",i)
    data=pd.read_csv("./"+s+i+names[0]+".csv")
    data.columns=["x","y"]
    plt.plot(data["x"], data["y"], label = f"r = {i}")# , kind="scatter")
    
plt.ylim(bottom=0,top=1) 
plt.xlim(left=0, right = 201)
    
plt.legend()
plt.show()
```
