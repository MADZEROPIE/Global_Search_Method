```python
import subprocess
str_of_r=["1.5","2.0","2.5","3.0","4.0"]

s="res"

for i in str_of_r:
    subprocess.run([r".\build\Release\builder_sample.exe",s+i,i])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

s="res"
names=["Hansen","Hill","Shekel"]

print(names[0]);
for i in str_of_r:
    print("r=",i)
    data=pd.read_csv("./"+s+i+names[0]+".csv")
    data.columns=["x","y"]
    data.plot(x="x", y="y" )#, kind="scatter")
    plt.ylim(bottom=0,top=1)
    plt.show()
```
